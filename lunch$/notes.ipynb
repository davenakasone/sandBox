## mortgages

```python
import matplotlib.pyplot as plt
import numpy as np
from ffuns import amortize_point

mrate   = 0.07 / 12    # monthly rate
nperiod = 12 * 30      # 30yr fixed
ploan   = 500.0e3      # amount of loan requested

interest = []
interest.append(0.0)
principal = []
principal.append(0.0)
balance = []
balance.append(ploan)

for nn in range(1, nperiod+1) :
    ipmt, ppmt, bal = amortize_point(ploan, mrate, nperiod, nn)
    interest.append(ipmt)
    principal.append(ppmt)
    balance.append(bal)

ntime = np.arange(0, nperiod+1)

fig1 = plt.figure()
plt.plot(ntime, balance)
plt.title("balance vs time")

fig2 = plt.figure()
plt.plot(ntime, interest)
plt.title("interest paid vs time")

fig3 = plt.figure()
plt.plot(ntime, principal)
plt.title("principal paid vs time")

plt.show()
```
