```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
```

```python
# the global matplotlib API...uses entire module, then invokes the instance (global calls modify)
x = np.arange(-10, 11)
plt.figure(figsize=(12, 6))
plt.title("y = +/- x**2")
plt.plot(x, x**2)
plt.plot(x, -1 * x**2)
```

```python
# no way to know what plot is being effected when you do this (multiple plots)
plt.figure(figsize=(12, 6))
plt.title("+/- x**2")

plt.subplot(1, 2, 1) # rows, cols, pannels
plt.plot(x, x**2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend('x**2', 'vertical')
plt.xlabel('x')
plt.ylabel('f(x)')

# implies switching the plot to draw on (order is important)

plt.subplot(1, 2, 2) # rows, cols, pannels
plt.plot(x, -1 * x**2)
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend('x**2', 'horizontal')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# use the oop method (explicit)
fig, axes = plt.subplots(figsize=(12, 6)) # gives you a figure with axes + reference to unique instance
axes.set_title("f(x) = +/- x**2")
axes.plot(
    x,
    x**2,
    color='red',
    linewidth=3,
    marker='o',
    markersize=8,
    label='x**2'
)
axes.plot(
    x,
    -1 * x**2,
    'b--',
    label='-x**2'
)
axes.set_xlabel('x')
axes.set_ylabel('f(x)')
axes.legend()
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
axes.plot(x, x+0, linestyle='solid')
axes.plot(x, x+1, linestyle='dashed')
axes.plot(x, x+2, linestyle='dashdot')
axes.plot(x, x+3, linestyle='dotted')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
axes.plot(x, x+0, '-og', label='solid green')
axes.plot(x, x+1, '--c', label='dashed cyan')
axes.plot(x, x+2, '-.b', label='dashed blue')
axes.plot(x, x+3, ':r', label='dotted red')
axes.legend()
```

```python
N = 50
x = np.random.randn(N)
y = np.random.randn(N)
colors = np.random.randn(N)
area = np.pi * (20 * np.random.randn(N))**2
plt.figure(figsize=(14,6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))
ax1 = fig.add_subplot(1, 2, 1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()
ax2 = fig.add_subplot(1, 2, 2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()
plt.show()
```

```python
valz = np.random.randn(1000)
plt.subplots(figsize=(12, 6))
plt.hist(valz, bins=100, alpha=0.8, histtype='bar', color='steelblue', edgecolor='green')
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

```python
density = scipy.stats.kde.gaussian_kde(valz)
plt.subplots(figsize=(12,6))
values = np.linspace(min(valz)-10, max(valz)+10, 100)
plt.plot(values, density(values), color='#FF7F00')
plt.fill_between(values, 0, density(values), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

```python
# combined
plt.subplots(figsize=(12, 6))
plt.hist(valz, bins=100, alpha=0.8, density=1, histtype='bar', color='steelblue', edgecolor='green')
plt.plot(values, density(values), color='#FF7F00', linewidth=3)
plt.xlim(xmin=-5, xmax=5)
plt.show()
```

```python
# bar plot
Y = np.random.rand(1,5)[0]
Y2 = np.random.rand(1,5)[0]
plt.figure(figsize=(12,4))
bw=0.5
plt.bar(np.arange(len(Y)), Y, width=bw, color='green')
```

```python
# stacked bar
plt.figure(figsize=(12, 6))
plt.bar(np.arange(len(Y)), Y, width=bw, color='green', label='Y')
plt.bar(np.arange(len(Y2)), Y2, width=bw*0.9, color='red', label='Y2')
plt.legend()
plt.show()
```
