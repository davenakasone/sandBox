```python
f = sin(x)
g = .5*cos(x)
display(f)
fx = lambdify(x,f,modules=['numpy'])
gx = lambdify(x,g,modules=['numpy'])
xVals = np.linspace(-math.pi,10,100)
plt.plot(xVals, fx(xVals), label = 'f(x) = sin(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = .5 * cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('graph on xy plane')
plt.legend()
plt.show()
```

```python
f = asin(x)
g = acos(x)
fx = lambdify(x,f,modules=['numpy']) # always good to express in lambdas
gx = lambdify(x,g,modules=['numpy'])
xVals = np.linspace(-1,1,100) # keep -1 to 1 or you get an error
plt.plot(xVals, fx(xVals), label = 'f(x) = asin(x)')
plt.plot(xVals, gx(xVals), label = 'g(x) = acos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('inverses')
plt.legend()
plt.show() 
```

```python
f = sp.ceiling(x) # smallest integer not less than x
g = sp.floor(x) # largest integer value not greater than x
h = x-g            # x-g #sp.frac(x) # in between part
display(f)
display(g)
display(h)
fx = lambdify(x,f,modules=['numpy']) 
gx = lambdify(x,g,modules=['numpy'])
hx = lambdify(x,h,modules=['numpy'])
xVals = np.linspace(-10,10,100)
plt.plot(xVals, fx(xVals), label = 'f(x) = ceiling')
plt.plot(xVals, gx(xVals), label = 'g(x) = floor')
plt.plot(xVals, hx(xVals), label = 'h(x) = fraction')
plt.xlabel('x')
plt.ylabel('y')
plt.title('functions')
plt.legend()
plt.show()
```
